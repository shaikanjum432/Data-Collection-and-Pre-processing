## Imbalanced Dataset

A dataset with an unequal class distribution

```python
# importing dependies
import pandas as pd
import numpy as np
```

```python
# Loading the dataset into Pandas dataframe
credit_card_df = pd.read_csv('/content/creditcard.csv')
```

```python
# First 5 rows of the dataframe
credit_card_df.head()
```

```python
#Checking how the labels are distributed in target column
credit_card_df['Class'].value_counts()
```

This is hightly imbalaced dataset

0 --> Legit Transactions

1 --> Fraudulent Transacions

### Seperating the legit and fraudulent transactions

```python
legit = credit_card_df[credit_card_df.Class == 0]
fraud = credit_card_df[credit_card_df.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

## Under-sampling

Building a sample dataset containing similar distribution of Legit and fraud transaction

Number of fraudulend transactions --> 778

```python
legit_sample = legit.sample(n = 778)
```

```python
print(legit_sample.shape)
```

### Concatinating two dataframes

```python
newcredit_df = pd.concat([legit_sample, fraud], axis = 0)
```

```python
newcredit_df['Class'].value_counts()
```

```python
# first 5 rows
newcredit_df.head()
```

```python
# last 5 rows
newcredit_df.tail()
```

First all the legit smaples labels will be placed and later all the fraudulent labels will be placed in the new dataframe
